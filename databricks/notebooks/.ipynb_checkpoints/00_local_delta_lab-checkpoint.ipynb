# PETROVA 300K — Local Delta Lab
**Runs locally in Jupyter + PySpark 3.5.3 + Delta Lake 3.2.1**

Covers: SparkSession, DataFrame API, Delta CRUD, Time Travel, Schema Evolution, Window Functions, MERGE (SCD2)

## 1. Initialize Spark + Delta Lake

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("PETROVA-Local-Lab") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog") \
    .getOrCreate()

print(f"Spark: {spark.version}")
print(f"Delta Lake ready")
```

## 2. Bronze — Raw Ingestion

```python
from pyspark.sql.types import StructType, StructField, StringType, DoubleType, TimestampType
from datetime import datetime

schema = StructType([
    StructField("equipment_id", StringType()),
    StructField("sensor_type", StringType()),
    StructField("reading_value", DoubleType()),
    StructField("status", StringType()),
    StructField("timestamp", TimestampType())
])

data = [
    ("EQ-001", "temperature", 72.5, "ACTIVE", datetime(2026, 5, 16, 10, 0)),
    ("EQ-001", "pressure", 14.7, "ACTIVE", datetime(2026, 5, 16, 10, 0)),
    ("EQ-002", "temperature", 185.3, "ACTIVE", datetime(2026, 5, 16, 10, 5)),
    ("EQ-002", "vibration", 0.08, "MAINTENANCE", datetime(2026, 5, 16, 10, 5)),
    ("EQ-003", "temperature", 95.1, "ACTIVE", datetime(2026, 5, 16, 10, 10)),
    ("EQ-003", "pressure", 15.2, "ACTIVE", datetime(2026, 5, 16, 10, 10)),
    ("EQ-004", "temperature", None, "INACTIVE", datetime(2026, 5, 16, 10, 15)),
    ("EQ-004", "vibration", 0.45, "ACTIVE", datetime(2026, 5, 16, 10, 15)),
]

df_bronze = spark.createDataFrame(data, schema)
df_bronze.show()
print(f"Records: {df_bronze.count()}")
```

```python
BRONZE_PATH = "/home/jovyan/data/bronze/sensor_readings"

df_bronze.write.format("delta").mode("overwrite").save(BRONZE_PATH)
print("Bronze layer written to Delta!")
```

## 3. Silver — Clean & Validate (3-Gate Pattern)

```python
from pyspark.sql import functions as F
from pyspark.sql.window import Window

bronze = spark.read.format("delta").load(BRONZE_PATH)

SILVER_PATH = "/home/jovyan/data/silver/sensor_cleaned"

df_silver = (bronze
    # Quality flag (Bronze Gate)
    .withColumn("quality_flag",
        F.when(F.col("reading_value").isNull(), "FAIL")
         .when(F.col("reading_value").between(0, 200), "PASS")
         .otherwise("WARN")
    )
    # Filter out INACTIVE
    .filter(F.col("status").isin("ACTIVE", "MAINTENANCE"))
    # Add derived columns
    .withColumn("reading_date", F.to_date("timestamp"))
    .withColumn("reading_hour", F.hour("timestamp"))
    # Deduplication
    .dropDuplicates(["equipment_id", "sensor_type", "timestamp"])
)

df_silver.show()
print(f"Bronze: {bronze.count()} → Silver: {df_silver.count()} (filtered {bronze.count() - df_silver.count()})")
```

```python
df_silver.write.format("delta").mode("overwrite").save(SILVER_PATH)
print("Silver layer written!")
```

## 4. Gold — Aggregated KPIs

```python
silver = spark.read.format("delta").load(SILVER_PATH)

GOLD_PATH = "/home/jovyan/data/gold/daily_sensor_kpi"

# Window function — rank readings per equipment
window_spec = Window.partitionBy("equipment_id").orderBy(F.desc("reading_value"))

df_gold = (silver
    .filter(F.col("quality_flag") == "PASS")
    .groupBy("equipment_id", "reading_date")
    .agg(
        F.count("*").alias("total_readings"),
        F.round(F.avg("reading_value"), 2).alias("avg_reading"),
        F.round(F.min("reading_value"), 2).alias("min_reading"),
        F.round(F.max("reading_value"), 2).alias("max_reading"),
        F.round(F.stddev("reading_value"), 2).alias("stddev_reading"),
        F.collect_set("sensor_type").alias("sensor_types")
    )
    .orderBy("equipment_id")
)

df_gold.show(truncate=False)
df_gold.write.format("delta").mode("overwrite").save(GOLD_PATH)
print("Gold layer written!")
```

## 5. Delta Lake — Time Travel

```python
from delta.tables import DeltaTable

dt = DeltaTable.forPath(spark, BRONZE_PATH)

# View history
dt.history().select("version", "timestamp", "operation", "operationMetrics").show(truncate=False)

# Read specific version
v0 = spark.read.format("delta").option("versionAsOf", 0).load(BRONZE_PATH)
print(f"Version 0 count: {v0.count()}")
```

## 6. Delta MERGE — Upsert (SCD Type 1)

```python
# New batch: 2 updates + 1 new record
new_data = [
    ("EQ-001", "temperature", 74.8, "ACTIVE", datetime(2026, 5, 16, 11, 0)),
    ("EQ-002", "temperature", 178.2, "ACTIVE", datetime(2026, 5, 16, 11, 0)),
    ("EQ-005", "temperature", 68.9, "ACTIVE", datetime(2026, 5, 16, 11, 0)),
]

df_new = spark.createDataFrame(new_data, schema)

dt_bronze = DeltaTable.forPath(spark, BRONZE_PATH)

# MERGE: update existing, insert new
(dt_bronze.alias("target")
    .merge(
        df_new.alias("source"),
        "target.equipment_id = source.equipment_id AND target.sensor_type = source.sensor_type"
    )
    .whenMatchedUpdateAll()
    .whenNotMatchedInsertAll()
    .execute()
)

# Verify
spark.read.format("delta").load(BRONZE_PATH).orderBy("equipment_id", "sensor_type").show()
print("MERGE complete — check version history:")
dt_bronze.history().select("version", "operation", "operationMetrics").show(truncate=False)
```

## 7. Schema Evolution

```python
from pyspark.sql.types import IntegerType

# New data with extra column
schema_v2 = StructType([
    StructField("equipment_id", StringType()),
    StructField("sensor_type", StringType()),
    StructField("reading_value", DoubleType()),
    StructField("status", StringType()),
    StructField("timestamp", TimestampType()),
    StructField("batch_id", IntegerType()),
])

new_with_col = [
    ("EQ-006", "temperature", 88.3, "ACTIVE", datetime(2026, 5, 16, 12, 0), 2),
]

df_evolved = spark.createDataFrame(new_with_col, schema_v2)

# mergeSchema = true allows adding new columns
df_evolved.write.format("delta") \
    .mode("append") \
    .option("mergeSchema", "true") \
    .save(BRONZE_PATH)

spark.read.format("delta").load(BRONZE_PATH).show()
print("Schema evolved — batch_id column added!")
```

## 8. Higher-Order Functions (Databricks DE exam topic)

```python
# Build array column for practice
df_arrays = (spark.read.format("delta").load(BRONZE_PATH)
    .groupBy("equipment_id")
    .agg(
        F.collect_list("reading_value").alias("all_readings"),
        F.collect_set("sensor_type").alias("sensor_types")
    )
)

df_arrays.createOrReplaceTempView("equipment_arrays")

# TRANSFORM — apply function to each element
spark.sql("""
    SELECT equipment_id,
           TRANSFORM(all_readings, x -> ROUND(x, 1)) AS rounded,
           FILTER(all_readings, x -> x IS NOT NULL AND x > 50) AS high_readings,
           SIZE(all_readings) AS total_count,
           EXISTS(all_readings, x -> x > 100) AS has_high
    FROM equipment_arrays
""").show(truncate=False)
```

## 9. Window Functions (SnowPro + Databricks exam topic)

```python
silver = spark.read.format("delta").load(SILVER_PATH)

w = Window.partitionBy("equipment_id").orderBy("timestamp")

df_window = (silver
    .withColumn("row_num", F.row_number().over(w))
    .withColumn("rank", F.rank().over(w))
    .withColumn("dense_rank", F.dense_rank().over(w))
    .withColumn("lag_value", F.lag("reading_value", 1).over(w))
    .withColumn("lead_value", F.lead("reading_value", 1).over(w))
    .withColumn("running_avg", F.round(F.avg("reading_value").over(w), 2))
)

df_window.select(
    "equipment_id", "sensor_type", "reading_value",
    "row_num", "rank", "lag_value", "lead_value", "running_avg"
).show()
```

## 10. Cleanup & Verify Full Pipeline

```python
print("=== PETROVA Medallion Pipeline Summary ===")
print(f"Bronze: {spark.read.format('delta').load(BRONZE_PATH).count()} records")
print(f"Silver: {spark.read.format('delta').load(SILVER_PATH).count()} records")
print(f"Gold:   {spark.read.format('delta').load(GOLD_PATH).count()} records")
print()
print("Delta versions:")
DeltaTable.forPath(spark, BRONZE_PATH).history().select("version", "operation").show()
print("\nLocal Delta Lab complete!")
```
